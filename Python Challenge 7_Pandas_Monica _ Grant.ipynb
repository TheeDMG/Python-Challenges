```python
import pandas as pd
```

```python
first_name = ('Dushane','Sully','Ranell','Gem','Lisa')
last_name= ('Hill','Light','Summerhouse', 'Marina', 'Summerhouse')
ages= (26, 25, 14, 13, 44)

print(first_name)
print(last_name)
print(ages)
```

```python
data = {'First Name': list(first_name), 'Last Name': list(last_name), 'Age': list(ages)}
```

```python
print(data)
```

```python
first_name_series = pd.Series(list(first_name)) 
last_name_series = pd.Series(list(last_name)) 
ages_series = pd.Series(list(ages))
```

```python
average_age = ages_series.mean()


print(f"The average employee age is {average_age} years.")
```

```python
ages_series.to_csv('ages.csv', index=False)
```
