## 1. Population by Country Line Chart

```python
df_country_by_population = df[['Country', 'Population']]
```

```python
df_country_by_population
```

```python
df_country_by_population.head(5) #first 5 countries in alpha-order 
```

```python
plt.figure(figsize=(10,3)) 
plt.plot(df_country_by_population.head(5)['Country'], df_country_by_population.head(5)['Population'], color='c', marker='D')
plt.xlabel('Country (First 5)')
plt.ylabel('Population (Millions)')
plt.title('Population by Country by Monica Grant')
plt.show()
```

## 2. GDP by Country Bar Chart

```python
df
```

```python
df
```

```python
df_country_gdp = df[['Country', 'GDP ($ per capita)']]
```

```python
df_country_gdp.tail(10)
```

```python
plt.figure(figsize=(10,3))
plt.bar(df_country_gdp.tail(5)['Country'], df_country_gdp.tail(5)['GDP ($ per capita)'], color='m')
plt.xlabel('Last 5 Countries')
plt.ylabel('GDP  ($ per capita)')
plt.title('GDP by Country Bar Chart by Monica Grant')
plt.show()
```
